<b>Problem #1:</b> Area of a square
>+ Given a square's width and height, display its area.
>+ The area of a sqare is its width multiplied by its height.
>+ Prompt the user to enter the required inputs.
>>use <code>input()</code> to prompt the user to enter the required values

<b>Problem #4:</b> Compute an employee's total weekly pay.

>+ Write a program that takes as inputs the hourly wage, total regular hours, and total overtime hours and displays an employee's total weekly pay.
>+ An employee's total weekly pay equals the hourly wage multiplied by the total number of regular hours plus any overtime pay.  Overtime pay equals the total overtime hours multiplied by 1.5 times the hourly wage.
>+ Use round() to round the output to 2 decimal places.
>>use <code>input()</code> to prompt the user to enter the required values

```python
# Request the inputs
hourly_wage = input('Please enter the hourly wage: ')
regular_hours = input('Please enter the regular working hours: ')
overtime_hours = input('Please enter the overtime working hours: ')
# Compute the result
weekly_pay = float(hourly_wage) * float(regular_hours) + float(overtime_hours) * float(hourly_wage) * 1.5
# Display the result
print('The weekly pay of this employee is', round(weekly_pay, 2))
```

<b>Problem #6:</b> Compute a person's income tax.
>+ Write a program that calculates a person's annual income tax.
>+ Note: as the example output reflects the values chosen by the Professor, your output values will likely be different. That will be fine.

1. Significant constants (hard-code these values in the program)
       tax rate
       standard deduction
       deduction per dependent
2. The inputs are:
       gross income 
       number of dependents
    >use <code>input()</code> to prompt the user to enter the input values
     
3. Computations:
       net income = gross income - the standard deduction - a deduction for each dependent 
       income tax = is a fixed percentage of the net income 
4. The outputs are
       the income tax, rounded to two figures


```python
# Initialize the constants
TAX_RATE = 0.1
STANDARD_DEDUCTION = 1000
DEDUCTION_PER_DEPENDENT = 300
# Prompt user for input values
gross_income = input('Enter the gross income: ')
dependent = input('Enter the number of dependents: ')
# Compute the income tax
net_income = float(gross_income) - STANDARD_DEDUCTION - DEDUCTION_PER_DEPENDENT * float(dependent)
income_tax = TAX_RATE * net_income
# Display the income tax
print('The income tax is $' + str(round(income_tax, 2)))
```

<b>Problem #9:</b> Try / Except / Else <br>

Modify the below program to satisy the following requirements:

>+ When a ValueError is raised, the program will display this message: Not a number!
>+ When an AssertionError is raised, the program will display this message: Not an even number!
>+ When no exception is raised, the program will display the reciprocal of the number entered.
>+ Your program must include the following keywords: assert, try, except, else.

Note: A ValueError is raised when a user types in data of an inappropriate type. <br>
Note: An AssertionError is raised when a user types in numeric data that is not even. <br>
Note: Use the modulus operator, sometimes also called the remainder operator, to determine whether an input number is either even or odd. <br>
Note: Do not include in the assert statement a condition; instead, however, 'catch' the exception using except.<br>
Note: Additional references on Python's assert statement: <br>
* https://stackoverflow.com/questions/5142418/what-is-the-use-of-assert-in-python


```python
try:
    num = int(input("Enter an even number: "))
    assert num % 2 == 0
except ValueError:
    print('Not a number!')
except AssertionError:
    print('Not an even number!')
else:
    print('The reciprocal of this number is', 1/num)
```
